### DETALHES TÉCNICOS DA BIBILITECA DE MACHINE LEARNING:
  - O Support Vector Machine (SVM) é um algoritmo de aprendizado de máquina supervisionado usado para classificação e regressão. Embora também digamos problemas de regressão, é mais adequado para classificação. 
  - O principal objetivo do algoritmo SVM é encontrar o hiperplano ótimo em um espaço N-dimensional que possa separar os pontos de dados em diferentes classes no espaço de recursos. 
  - O hiperplano tenta que a margem entre os pontos mais próximos de diferentes classes seja a máxima possível. A dimensão do hiperplano depende do número de feições. Se o número de feições de entrada for dois, então o hiperplano é apenas uma linha. 
  - Se o número de recursos de entrada for três, o hiperplano se tornará um plano 2-D. Torna-se difícil imaginar quando o número de recursos excede três. 
  https://www.geeksforgeeks.org/support-vector-machine-algorithm/

### Criação de uma função para detalhamento dos dados Null ( data null ) e Types
  - Criarei duas funções e analisarei qual melhor se encaixa para os processos

```python
# Função para mapear dos dados do dataset 

def summary_missing_values(data):
    missing_values = pd.DataFrame(df_prostate_cancer.isna().sum()/df_prostate_cancer.shape[0])
    missing_values.reset_index(drop=True, inplace=True)    
    missing_values.columns = ["Absoluto dados ausentes(%)"]
    missing_values["Column"] = pd.Series(df_prostate_cancer.columns).values
    # Make index be the column "Absolute Missing (%)"
    missing_values.reset_index(drop=True, inplace=True)
    missing_values.set_index("Column", inplace=True)
    
    # Calculando os valores comparando com o tatal de dados por coluna e os dados ausentes quantitativos em percentual
    column_missing = []
    column_total   = []
    column_miss    = []
    column_type    = []
    
    for column in df_prostate_cancer.columns:
        col_miss = df_prostate_cancer[column].isnull().sum()
        total_values = df_prostate_cancer[column].count()       
        column_miss.append(col_miss)
        column_total.append(total_values) 
        column_missing.append((col_miss/total_values))
        column_type.append(column_type)
        
    # Construindo o dataframe que receberá os valores
    missing_values["Perc. Abs. ausentes(%)"] = pd.Series(column_missing).values
    missing_values["Perc. da equivalência(%)"] = -(pd.Series(column_missing).values-1)
    missing_values["Valores únicos"] = pd.Series(column_total).values
    missing_values["Valores absolutos"] = pd.Series(column_miss).values
    missing_values['Variáveis tipos'] = pd.DataFrame(df_prostate_cancer.dtypes, columns=['column_type'])
    missing_values.sort_values(by="Absoluto dados ausentes(%)", ascending=False, inplace=True)
    missing_columns = list(missing_values.query("`Absoluto dados ausentes(%)` > 0").index)
    
    pd.options.display.float_format = '{:.2%}'.format
    return missing_values, missing_columns

missing_values, missing_columns_names = summary_missing_values(df_prostate_cancer)
missing_values
```

### Descritiva das distribuíções com metodos <em>***skew e kurtosis***<em>
  - Métricas de média e mediana, minimo e máximo e **<em>centrais<em>** com skew e kurtosis descritivas para entendimento das distribuções numéricas  
    - Aplicação para séries temporais com ARIMA

### Format para retirar os caracteres indezejados 


```python
# Incluindo somente variáveis numéricas para uma análise descritiva com métricas (dados estatísticos, algebra linear com numpy)

# df_prostate_cancer.columns=df_prostate_cancer.columns.str.replace('[#,@,&,e,+]','') 
pd.options.display.float_format = '{:.4}'.format
num_attributes = df_prostate_cancer.select_dtypes( include=[ 'int64', 'float64' ] )
# cat_attributes = df_kc.select_dtypes( exclude=[ 'int64', 'float64' ] )

# Slice da coluna (id), mas não inplace
num_attributes = num_attributes.iloc[:, : ]
```

```python
# Aplicação dos métodos numpy para produzir métricas centrais e dispersões da medida central de cadas variável
# coef_var = (statistics.stdev(df_light.grupo3) / len(df_light.grupo3)) * 100 
# print(f'Coeficiente de Variação é:\t {coef_var:.1f}%') # Coeficiente de variação

tcd1 = pd.DataFrame( num_attributes.apply( np.mean ) ).T 
tcd2 = pd.DataFrame( num_attributes.apply( np.median ) ).T
tcd3 = pd.DataFrame( num_attributes.apply( np.var ) ).T 

# Dimensão - std, min, max, range, skew, kurtosis
dm1 = pd.DataFrame( num_attributes.apply( np.std )).T 
dm2 = pd.DataFrame( num_attributes.apply( min )).T 
dm3 = pd.DataFrame( num_attributes.apply( max )).T
dm4 = pd.DataFrame( num_attributes.apply( lambda x: x.max() - x.min())).T
dm5 = pd.DataFrame( num_attributes.apply( lambda x: x.skew())).T  
dm6 = pd.DataFrame( num_attributes.apply( lambda x: x.kurtosis())).T

# Concatenando as métricas dos Data Frames para gerar um novo dataframe e fazer a concatenação das métricas para display.
estatistica_descritiva = pd.concat([dm2, dm3, dm4, tcd1, tcd2, tcd3, dm1, dm5, dm6]).T.reset_index() 
estatistica_descritiva.columns = ['ATRIBUTOS', 'MIN', 'MAX', 'RANGE', 'MEDIA', 'MEDIANA', 'VARIÂNCIA', 'STD', 'SKEW_CENTRAL', 'KURTOSIS']

print('')
print('DESCRITIVO ESTATÍSTICO COM MÉTODOS SKEW E KURTOSIS DAS MÉTRICAS DE DISPERSÃO E CENTRAIS'.center(95))
print("")

# Imprimindo o dataset concatenado para ser consumido
estatistica_descritiva[estatistica_descritiva !=~0].round(4)
```

```python
### Function 
'''
  Dbian Data Scientist 
  Função responsável por anotar os indicadores das barras no gráfico.
  Entrada: tipo do gráfico ('bar' ou 'barh')
'''
def autolabel(tipo):
  if tipo == 'bar':
    for bar in plot.patches:
      plot.annotate(format(bar.get_height()),
                xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                xytext=(0, 8),
                textcoords='offset points',
                ha='center', 
                va='center',
                size=11)
  elif tipo == 'barh':
    for bar in plot.patches:
      plot.annotate(format(bar.get_width()),
                  xy = (bar.get_width() + 0.3, bar.get_y() + 0.38),
                  xytext=(15, -2),
                  textcoords='offset points',
                  ha='center', 
                  va='center',
                  size=11)
```

### Tratamentos e processos para implentar Machile Learning - AI
  - Nesse momento separo os dados de treino e testes para dar início aos trabalhos de predição.
   - Antes da separação, guardo uma cópia e passo para outro objeto python "df_predict".
   > Para esse projeto, depois de fazer o split dos dados, removerei os outliers para trabalhar de forma mais assertiva e validar algumas percepções

```python
df_predict=df_prostate_cancer.copy()
```

```python
# Abrindo as colunas para entendimento do processo

df_predict.columns
```

```python
# Separação dos dados em Treino e Testes em X,y

train_x, test_x, train_y, test_y = train_test_split(
      df_predict.drop(['id','diagnosis_result'], axis=1), 
      df_predict[['diagnosis_result']], 
      test_size=0.30,
      random_state=41)

print(f'QUANTITATIVO DAS AMOSTRAS:\n')
print(f'Dados de treino em x:\t', train_x.shape)
print(f'Dados de teste em x:\t', test_x.shape)
```

___

### Processamento e fazendo ajustes de Outliers
  - Outliers são valores muito discrepandates se comparados atravéz de uma média ponderada com os demais valores e para aplicação de Machile learning, preciso tratar para que os dados de treino e testes possam ser mais assertivos.

```python
# Passando um "for" para cada linha afim de gerar a remoção dos outliers dos dados de treino e testes da variável "diagnosis_result".

for column in train_x.columns:
    
    df_train_1 = train_x[(train_y.diagnosis_result=='B') & (
                 train_x[column]<np.mean(train_x.loc[train_y.diagnosis_result=='B',column])+3*np.std(train_x.loc[train_y.diagnosis_result=='B',column]))]
    df_test_1 = test_x[(test_y.diagnosis_result=='B') & (
                 test_x[column]<np.mean(train_x.loc[train_y.diagnosis_result=='B',column])+3*np.std(train_x.loc[train_y.diagnosis_result=='B',column]))]
    
    label_train_1 = train_y[(train_y.diagnosis_result=='B') & (
                 train_x[column]<np.mean(train_x.loc[train_y.diagnosis_result=='B',column])+3*np.std(train_x.loc[train_y.diagnosis_result=='B',column]))]
    label_test_1 = test_y[(test_y.diagnosis_result=='B') & (
                 test_x[column]<np.mean(train_x.loc[train_y.diagnosis_result=='B',column])+3*np.std(train_x.loc[train_y.diagnosis_result=='B',column]))]
    
    df_train_2 = train_x[(train_y.diagnosis_result=='M') & (
                 train_x[column]<np.mean(train_x.loc[train_y.diagnosis_result=='M',column])+3*np.std(train_x.loc[train_y.diagnosis_result=='M',column]))]
    df_test_2 = test_y[(test_y.diagnosis_result=='M') & (
                 test_x[column]<np.mean(train_x.loc[train_y.diagnosis_result=='M',column])+3*np.std(train_x.loc[train_y.diagnosis_result=='M',column]))]
    
    label_train_2 = train_y[(train_y.diagnosis_result=='M') & (
                 train_x[column]<np.mean(train_x.loc[train_y.diagnosis_result=='M',column])+3*np.std(train_x.loc[train_y.diagnosis_result=='M',column]))]
    label_test_2 = test_y[(test_y.diagnosis_result=='M') & (
                 test_x[column]<np.mean(train_x.loc[train_y.diagnosis_result=='M',column])+3*np.std(train_x.loc[train_y.diagnosis_result=='M',column]))]    
```

### Aplicação de concatenação dos dataframes criados acima.
  - Gerando os quantitativos de cada parcição dados de treino e testes em x_ e y_

```python
# Concatenando os dados com os ajustes feitos de remoção dos outliers

train_x = pd.concat([df_train_1,df_train_2])
train_y = pd.concat([label_train_1,label_train_2])

test_y = pd.concat([df_test_1,df_test_2])
test_y = pd.concat([label_test_1,label_test_2])

print(f'QUANTITATIVO DAS AMOSTRAS:\n')
print(f'Dados de treino em x:\t', train_x.shape)
print(f'Dados de teste em x:\t', test_x.shape)
```

### Criação de uma function (função) 
  >- Criarei uma função para preparar os dados de treino e teste para aplicação dos MODELOS 

```python
def modelo_matrix(matrix,pred):
    class_names= [0,1]# name  of classes
    fig, ax = plt.subplots()
    tick_marks = np.arange(len(class_names))
    plt.xticks(tick_marks, class_names)
    plt.yticks(tick_marks, class_names)
    # create heatmap
    sns.heatmap(pd.DataFrame(matrix), annot=True, cmap="YlGnBu" ,fmt='g')
    ax.xaxis.set_label_position("top")
    plt.tight_layout()
    plt.title('Confusion matrix', y=1.1)
    plt.ylabel('Actual label')

    plt.tight_layout()
```

```python
# Efetuando a predição das classes com o modelo Naive Bayes e gerando a matriz de confusão.

pred_y = nb.predict(test_x)
cnf_matrix = metrics.confusion_matrix(pred_y,test_y)
modelo_matrix(cnf_matrix,test_y)

# Cálculo e predição do Modelo 
relatorio = classification_report(pred_y,test_y)
print(f'\nREPORT METRICS CONFUSION MATRIZ\n\n',relatorio)
print('\nA acuracia do modelo Naive Bayes é:\t\t{:.2f}%'.format(acuracia))
```

___
### Modelo 3 - SUPORT VECTOR MACHINE (SVM)

```python
#  Efetuando a predição das classes com o modelo Gaussian Naive Bayes e gerando a matriz de confusão. 

svm = SVC(random_state = 1)
svm.fit(train_x, train_y.values.ravel())

pred_y=svm.predict(test_x)
acuracia = metrics.accuracy_score(pred_y,test_y.values.ravel())*100

print(f'\nGERANDO O SCORE E ACURÁCIA DO MODELO SVM\n')
print('A ACURACIA DO SVM Algorithm é:{:.2f}%'.format(acuracia))
```

```python
# Efetuando a predição das classes com o modelo Gaussian Naive Bayes e gerando a matriz de confusão.

pred_y = svm.predict(test_x)
cnf_matrix = metrics.confusion_matrix(pred_y,test_y)
modelo_matrix(cnf_matrix,test_y)

# Cálculo e predição do Modelo 
relatorio = classification_report(pred_y,test_y)
print(f'\nREPORT METRICS CONFUSION MATRIZ\n\n',relatorio)
print('\nA acuracia do modelo Naive Bayes é:\t\t {:.2f}%'.format(acuracia))
```

___
### Modelo 4 - KNN 

```python
# Efetuando a predição das classes com o modelo Gaussian KNN e gerando a matriz de confusão. 
# Buscando a melhor métrica 

score = []

for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors = i) #n_neighbors means k
    knn.fit(train_x, train_y.values.ravel())
    score.append(knn.score(test_x, test_y.values.ravel()))
    
plt.plot(range(1,20), score)
plt.xticks(np.arange(1,20,1))
plt.xlabel('K neighbors')
plt.ylabel('Score')

gca()
plt.tight_layout()

acuracia = max(score)*100
print(f'\nGERANDO O SCORE E ACURÁCIA DO MODELO KNN\n')
print('\nMétrica máxima KNN Score é:\t{:.2f}%'.format(acuracia))
print('A acuracia do Score é:\t\t{:.2f}%'.format(acuracia))
```

### Modelo 5 - Regressão Logistica

```python
# Efetuando a predição das classes com o modelo Regressão Logistica e gerando a matriz de confusão.  

logreg = LogisticRegression(max_iter=30)
logreg.fit(train_x, train_y.values.ravel())
pred_y=logreg.predict(test_x)
acuracia = metrics.accuracy_score(pred_y,test_y.values.ravel())*100 
print(f'\nGERANDO O SCORE E ACURÁCIA DO MODELO REGRESSÃO LOGÍSTICA\n')
print('Teste de acuracia do modelo RL é:{:.2f}%'.format(acuracia))
```

### Modelo 6 - Neural Network
  - Apartir desssa nova versão, nao usa-se mais init passando o init para kernel_inicializer=

  - https://keras.io/api/layers/core_layers/

```python
# Efetuando a predição das classes com o modelo Neural Network (KERAS) e gerando a matriz de confusão. 

model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))

# Compilando o sequenciaql model do KERAS
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 

# Criando o fit do KERAS no dataset de treinamento e teste da
model.fit(train_x, train_y.replace({'B':0,'M':1}), epochs=100, batch_size=8)

# Avaliação do Moelo KERAS Rede Neural 
_,acuracia = model.evaluate(test_x, test_y.replace({'B':0,'M':1}))

print(f'\nACURÁCIA DO KERAS E SCORE')
print('\nA acuracia e Score do modelo Keras é:\t\t\t\t\t\t{:.2f}%'.format(acuracia))
```

```python
# Efetuando a predição das classes com o modelo KNN e gerando a matriz de confusão. 

pred_y = model.predict(test_x)
pred_y = np.round(pred_y)

# Cálculo e predição do Modelo
relatorio = classification_report(pred_y,test_y.replace({'B':0,'M':1}))

print(f'\nREPORT METRICS CONFUSION MATRIZ\n\n',relatorio)
print('O score do modelo Regressão Logistica é:\t{:.2f}%'.format(acuracia))
```
